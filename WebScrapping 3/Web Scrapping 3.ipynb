```python
URL = []



start=0
end=3

for page in range(start,end):
    
     #URL
    url=driver.find_elements(By.XPATH, '//a[@class="a-link-normal s-no-outline"]')
    for i in url:
        U = (i.get_attribute('href'))
        if U == (""):
            URL.append("-")
        else:
            URL.append(U)
            

   
        
    
    
    

    
    
```

```python
print (len(URL))
```

```python

Brand = []
P_name = []
RnE = []
Avl = []
Price = []
Del = []
       



for i in URL:
    driver.get(i)
    brand=driver.find_elements(By.XPATH, '//tr[@class="a-spacing-small po-brand"]')
    for x in brand:
        b = (x.text)
        if b == (""):
            Brand.append('-')
        else:
            Brand.append(b)
            
    p_name=driver.find_elements(By.XPATH, '//span[@class="a-size-large product-title-word-break"]')
    for x in p_name:
        pn = (x.text)
        if pn == (""):
            P_name.append('-')
        else:
            P_name.append(pn)
                
    price=driver.find_elements(By.ID,("corePriceDisplay_desktop_feature_div"))
    for i in price:
        pr = (i.text)
        if pr != (""):
            Price.append(pr)
        
    delv=driver.find_elements(By.ID ,"mir-layout-DELIVERY_BLOCK-slot-PRIMARY_DELIVERY_MESSAGE_LARGE")
    for x in delv:
        d = (x.text)
        if d == (""):
            Del.append("-")
        else:
            Del.append(d)
   
    
    
                
    
    
    
    
    rne=driver.find_elements(By.ID, "RETURNS_POLICY")
    for x in rne:
        re = (x.text)
        if re == (""):
            RnE.append("-")
        else:
            RnE.append(re)
                
                

                
    avl=driver.find_elements(By.XPATH, '//span[@class="a-size-medium a-color-success"]')
    for x in avl:
        a = (x.text)
        if a == (""):
            Avl.append("-")
        else:
            Avl.append(a)       

                
```

```python
print (len(P_name), len(Brand),len(Avl), len(RnE),len(Price),len(Del))
```

```python
df = pd.DataFrame ({'Brand':Brand,'Product Name':P_name,'Price':Price,'Availability':Avl,'Delivery':Del,'URL':URL})
df
```

```python
#Libraries
import pandas as pd
import selenium
from bs4 import BeautifulSoup
import time
import requests

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import StaleElementReferenceException , NoSuchElementException

import warnings
warnings.filterwarnings ('ignore')
```

```python
driver = webdriver.Chrome(r"C:\Users\pooja\chromedriver.exe")
```

```python
driver.get('https://www.youtube.com/watch?v=2hl-H3ygx_I')
```

```python
Comments = []
Time = []
Likes = []

comments=driver.find_elements(By.XPATH, '//yt-formatted-string[@class="style-scope ytd-comment-renderer"]')
for x in comments:
    r = (x.text)
    if r == (""):
        Comments.append('-')
    else:
        Comments.append(r)
        
time=driver.find_elements(By.XPATH, '//yt-formatted-string[@class="published-time-text style-scope ytd-comment-renderer"]')
for x in time:
    r = (x.text)
    if r == (""):
        Time.append('-')
    else:
        Time.append(r)

likes=driver.find_elements(By.ID, "vote-count-middle")
for x in likes:
    r = (x.text)
    if r == (""):
        Likes.append('-')
    else:
        Likes.append(r)
        
Likes
```

```python
print (len(Comments),len(Time),len(Likes))
```

```python
df = pd.DataFrame({'Comments':Comments,'Time':Time,'Likes':Likes})
df
```
